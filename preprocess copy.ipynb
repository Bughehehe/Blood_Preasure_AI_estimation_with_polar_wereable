```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from text file
data_PPG = pd.read_csv("./polar3/Polar_Sense_9F124324_20240425_222851_PPG_fixed.txt", delimiter=";")

# Convert timestamp columns to datetime objects
data_PPG['Phone timestamp'] = pd.to_datetime(data_PPG['Phone timestamp'], format='%Y-%m-%dT%H:%M:%S.%f')

# Obliczenie redniej wartoci sygnau
mean_channel_0 = data_PPG['channel 0'].mean()
mean_channel_ambient = data_PPG['ambient'].mean()

# Odjcie redniej wartoci od obu serii danych
data_PPG['channel 0'] -= mean_channel_0
data_PPG['channel 0 - ambient'] = data_PPG['channel 0'] - data_PPG['ambient'] + mean_channel_ambient

# Plot
plt.figure(figsize=(10, 6))  # Ustawiamy rozmiar caego wykresu

# Plot 1: channel 0
plt.plot(data_PPG['Phone timestamp'], data_PPG['channel 0'], marker='o', linestyle='-', label='Channel 0')

# Plot 2: channel 0 - ambient
plt.plot(data_PPG['Phone timestamp'], data_PPG['channel 0 - ambient'], marker='o', linestyle='-', label='Channel 0 - Ambient')

plt.title('Channel 0 and Channel 0 - Ambient over Time')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.grid(True)
plt.legend()  # Dodajemy legend
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# Function to apply Butterworth high-pass filter
def butter_highpass_filter(data, cutoff, fs, order=5):
    nyquist = 0.5 * fs
    normal_cutoff = cutoff / nyquist
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    y = filtfilt(b, a, data)
    return y

# Read data from text file
data_PPG = pd.read_csv("./polar3/Polar_Sense_9F124324_20240425_222851_PPG_fixed.txt", delimiter=";")

# Convert timestamp columns to datetime objects
data_PPG['Phone timestamp'] = pd.to_datetime(data_PPG['Phone timestamp'], format='%Y-%m-%dT%H:%M:%S.%f')

data_PPG['channel 0'] = data_PPG['channel 0'] - data_PPG['ambient']
# Calculate mean signal value
mean_channel_0 = data_PPG['channel 0'].mean()

# Subtract mean value from channel 0
data_PPG['channel 0'] -= mean_channel_0


# Apply high-pass filter to channel 0
fs = 55  # Sampling frequency (Hz)
cutoff = 1  # High cutoff frequency (Hz) - example value, you can adjust it according to your needs
data_PPG['filtered_channel'] = butter_highpass_filter(data_PPG['channel 0'], cutoff, fs)

# Plot
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot original channel 0 and filtered channel
plt.plot(data_PPG['Phone timestamp'], data_PPG['channel 0'], label='Original Channel 0')
plt.plot(data_PPG['Phone timestamp'], data_PPG['filtered_channel'], label='Filtered Channel', linestyle='--')

plt.title('Original Channel 0 vs Filtered Channel (High-pass Filter at 55 Hz)')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

# Adding a zoomed-in plot for the filtered signal
plt.figure(figsize=(10, 3))  # Set the figure size for the zoomed-in plot
plt.plot(data_PPG['Phone timestamp'], data_PPG['filtered_channel'], label='Filtered Channel', linestyle='-', color='orange')

plt.title('Zoomed-in Filtered Channel')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.xlim(data_PPG['Phone timestamp'].iloc[0], data_PPG['Phone timestamp'].iloc[-1])  # Set x-axis limits to match the original plot
plt.grid(True)
plt.legend()

plt.show()
```
