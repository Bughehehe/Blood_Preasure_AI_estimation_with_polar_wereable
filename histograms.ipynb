```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych z pliku tekstowego
df = pd.read_csv('pressure.txt', delimiter=' ', header=None, names=['Filename', 'SYS', 'DIA'])

# Tworzenie histogramu dla SYS
plt.figure(figsize=(10, 5))
n, bins, patches = plt.hist(df['SYS'], bins=10, color='blue', edgecolor='black', alpha=0.7, rwidth=0.9)
plt.title('Histogram dla iloci zebranych prbek cinienia skurczowego')
plt.xlabel('Cinienie skurczowe [mmHg]')
plt.ylabel('Ilo prbek [-]')
plt.grid(axis='y', linestyle='--')
plt.yticks(range(int(max(n)) + 1))

# Dostosowanie etykiet na osi X
bin_edges = [int(b) for b in bins]
bin_labels = [f'{bin_edges[i]}-{bin_edges[i+1]-1}' for i in range(len(bin_edges)-1)]
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, bin_labels, rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Tworzenie histogramu dla DIA
plt.figure(figsize=(10, 5))
n, bins, patches = plt.hist(df['DIA'], bins=10, color='green', edgecolor='black', alpha=0.7, rwidth=0.9)
plt.title('Histogram dla iloci zebranych prbek cinienia rozkurczowego')
plt.xlabel('Cinienie rozkurczowe [mmHg]')
plt.ylabel('Ilo prbek [-]')
plt.grid(axis='y', linestyle='--')
plt.yticks(range(int(max(n)) + 1))

# Dostosowanie etykiet na osi X
bin_edges = [int(b) for b in bins]
bin_labels = [f'{bin_edges[i]}-{bin_edges[i+1]-1}' for i in range(len(bin_edges)-1)]
bin_centers = 0.5 * (bins[:-1] + bins[1:])
plt.xticks(bin_centers, bin_labels, rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytanie danych z pliku tekstowego
df = pd.read_csv('pressure.txt', delimiter=' ', header=None, names=['Filename', 'SYS', 'DIA'])

# Tworzenie histogramu dla SYS
plt.figure(figsize=(10, 5))
unique_sys_values = sorted(df['SYS'].unique())
plt.hist(df['SYS'], bins=len(unique_sys_values), color='blue', edgecolor='black', alpha=0.7, rwidth=0.9)
plt.title('Histogram for SYS')
plt.xlabel('SYS Value')
plt.ylabel('Number of Samples')
plt.grid(axis='y', linestyle='--')
plt.xticks(unique_sys_values, rotation=45, ha='right')
plt.yticks(range(int(df['SYS'].value_counts().max()) + 1))
plt.tight_layout()
plt.show()

# Tworzenie histogramu dla DIA
plt.figure(figsize=(10, 5))
unique_dia_values = sorted(df['DIA'].unique())
plt.hist(df['DIA'], bins=len(unique_dia_values), color='green', edgecolor='black', alpha=0.7, rwidth=0.9)
plt.title('Histogram for DIA')
plt.xlabel('DIA Value')
plt.ylabel('Number of Samples')
plt.grid(axis='y', linestyle='--')
plt.xticks(unique_dia_values, rotation=45, ha='right')
plt.yticks(range(int(df['DIA'].value_counts().max()) + 1))
plt.tight_layout()
plt.show()
```
